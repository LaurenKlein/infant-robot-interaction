```python
import pandas as pd
import datetime
import os
```

```python
df_path = "NDF/"
state_path = "States/"
save_path = "Affect/"

offset_dict = {'2':'00:00:45:10',
               '3':'00:03:53:66',
               '4':'00:01:14:84',
               '6':'00:00:08:11',
               '7':'00:00:00:00',
               '8':'00:00:49:52',
               '9':'00:00:12:32',
               '12':'00:00:41:11',
               '14':'00:03:22:20'
              }
```

```python
def sync(a):
    a = datetime.timedelta(hours=a.hour, minutes=a.minute, seconds=a.second, microseconds=a.microsecond)
    b = a + datetime.timedelta(hours=offset.hour, minutes=offset.minute, seconds=offset.second, microseconds=offset.microsecond)
    return b

def dt_td(temp):
    temp = datetime.timedelta(hours=temp.hour, minutes=temp.minute, seconds=temp.second, microseconds=temp.microsecond)
    return temp

def func(x, start_time_col, end_time_col, label_col):
    for i in range(len(code)):
        if x < code[start_time_col].iloc[0] or x > code[end_time_col].iloc[len(code)-1]:
            return 'n'
        if x <= code[end_time_col].iloc[i]:
            return code[label_col].iloc[i]
    return 'n'
```

```python
for df_name in os.listdir(df_path):
    study_num = df_name.split("_")[1][2:]
    print("Processing infant: {}".format(study_num))
    state_study_num = [state_file_name.split('_')[0][4:] for state_file_name in os.listdir(state_path)]
    state_file_name = os.listdir(state_path)[state_study_num.index(study_num)]
    
    df_file_path = os.path.join(df_path, df_name)
    state_file_path = os.path.join(state_path, state_file_name)
    
    df = pd.read_csv(df_file_path)
    
    code = pd.read_csv(state_file_path, sep="\t", header=None)
    code = code.drop(columns=[1]) #NaN
    code = code.drop(index=code[code[0]=='condition'].index) #movement
    code = code.drop(index=code[code[0]=='movement'].index)
    code = code.reset_index(drop=True)
    
    if len(code.columns) == 8:
        start_time_col = 2
        end_time_col = 4
        label_col = 8
    
    elif len(code.columns) == 5:
        start_time_col = 2
        end_time_col = 3
        label_col = 5

    offset = offset_dict[str(study_num)] #change
    offset = pd.to_datetime(pd.Series([offset]), format= '%H:%M:%S:%f' ).dt.time[0]
    
    df['timestamp'] = pd.to_datetime(df['timestamp'], format= '%H:%M:%S:%f').dt.time
    code[start_time_col] = pd.to_datetime(code[start_time_col], format= '%H:%M:%S.%f').dt.time
    code[end_time_col] = pd.to_datetime(code[end_time_col], format= '%H:%M:%S.%f').dt.time
    
    df['synced_timestamp'] = df['timestamp'].apply(sync)
    code[start_time_col] = code[start_time_col].apply(dt_td)
    code[end_time_col] = code[end_time_col].apply(dt_td)
    
    df['affect'] = df['synced_timestamp'].apply(func, args=(start_time_col, end_time_col, label_col))
    
    df.to_csv(save_path + '/TD' + str(study_num) + '_Affect.csv', index=False)
```
