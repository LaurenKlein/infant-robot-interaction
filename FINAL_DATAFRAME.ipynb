```python
import numpy as np
import pandas as pd
import ast
```

```python
df_baby = pd.read_csv('NSF3_TD14.csv')
```

# Euclidean Distances 

```python
col_names1 =  [None]*2279
col_names1[0] = 'timestamp'

i = 1
while i<=2278:
    col_names1[i] = 'feature'+str(i)
    i += 1
    
df_dist  = pd.DataFrame(columns = col_names1)
```

```python
df_dist
```

```python
for ts_index in range(len(df_baby['timestamp'])):
    distances = []
    lm = np.array(ast.literal_eval(df_baby['68 landmarks'][ts_index]))
    #normalize
    right_eye_lm = lm[39]
    left_eye_lm = lm[42]
    dist = np.linalg.norm(right_eye_lm - left_eye_lm)
    for i in range(68):
        a = np.array(lm[i])
        distance = np.linalg.norm(lm[i+1:] - a, axis=1)
        #normalized distance
        distance = np.divide(distance, dist)
        distances.append(list(distance))
    flat_distances = [item for sublist in distances for item in sublist]
    flat_distances.insert(0, df_baby['timestamp'][ts_index])
    df_dist.loc[len(df_dist)] = flat_distances
    print("Done with " + str(ts_index) + " out of " + str(len(df_baby['timestamp'])))
```

```python
df_dist
```

# Differences

```python
df_diff = df_dist.drop(columns=['timestamp'])
```

```python
df_diff = df_diff.diff()
```

```python
df_diff
```

```python
df_diff.iloc[[0]] = [0]
```

```python
df_diff
```

# Combine

```python
result = pd.concat([df_dist, df_diff], axis=1, sort=False)
```

```python
result
```

```python
result.to_csv('NSF3_TD14_Norm_DataFrame.csv', index=None)
```
